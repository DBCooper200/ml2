```python
import pandas as pd
from sklearn import linear_model
```

```python
from word2number import w2n
```

```python
df=pd.read_csv('hiring.csv')
```

```python
df
```

```python
df.experience=df.experience.fillna("zero")
```

```python
df
```

```python
df.experience[0]
```

```python
num_array=[]
for i in range(len(df.experience)):
  num_array.append(w2n.word_to_num(df.experience[i]))
df.experience.replace(num_array,inplace=True)
```

```python
df['experience'] = num_array
```

```python
df.experience

```

```python
df
```

```python
new_column_names = {'test_score(out of 10)': 'test_score','interview_score(out of 10)': 'inter_score','salary($)':'sal'}
df.rename(columns=new_column_names, inplace=True)
```

```python
df
```

```python
df.test_score=df.test_score.fillna(df.test_score.median())
```

```python
df
```

```python
model=linear_model.LinearRegression()
```

```python
model=model.fit(df[['experience','test_score','inter_score']],df.sal)
```

```python
model.predict([[2,9,6]])
```

```python
model.predict([[12,10,10]])
```
